```python
#=======read data from web and write csv=====================
from selenium import webdriver
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as Req
import pandas as pd
import re
import csv
import xlsxwriter
my_url="https://pythonbricks.pythonanywhere.com/covid-spread/track-filter-country/"
response=Req(my_url)
page_html=response.read()
page_soup=soup(page_html,"html.parser")
plaintext = page_html.decode('utf8')
links = re.findall("window.location=[\"\'](.*?)[\"\']", plaintext)

#print(links)
#create csv file 
with open("covid19_csv.csv", "w", newline='') as f:
    header = ['Country','Confirmed Cases', 'Active Cases', 'Recovered Cases','Fatal Cases','Death Rate','Recovery Rate','Last Updated']
    writer = csv.writer(f, delimiter=',')
    writer.writerow(header)
    
    
    for i in links:
        data_list=[]
        ur="https://pythonbricks.pythonanywhere.com/"+i
        response1 =Req(ur)
        countries_page=response1.read()
        countries_soup=soup(countries_page,"html.parser")



        for countries_rows in countries_soup.findAll("div" ,{"class":"card"}):
            for countries_row in countries_rows.findAll("b"):
                data=countries_row.text
                data_list.append(data)
        writer.writerow(data_list)
           
```

```python
#======================read csv and write country wise worksheets========
from csv import *
import xlsxwriter
#create Excelsheet

df=pd.read_csv("covid19_csv.csv")
df["Confirmed Cases"]=df["Confirmed Cases"].str.replace(",","").astype(float)
df["Active Cases"]=df["Active Cases"].str.replace(",","").astype(float)
df["Recovered Cases"]=df["Recovered Cases"].str.replace(",","").astype(float)
df["Fatal Cases"]=df["Fatal Cases"].str.replace(",","").astype(float)
#print(df)
#workbook = xlsxwriter.Workbook('covid19.xlsx')
with pd.ExcelWriter('covid19.xlsx') as writer:

#worksheet = workbook.add_worksheet(countryname)
    for index,row in df.iterrows():
        #print(row)
        #sheet_name = "{:30.30}".format(row[0])
        sheet_name = row[0].replace("_","")
        #print (sheet_name)
        row.to_excel(writer,sheet_name=sheet_name)

    


#workbook.close()

   
```
